```python
start = 0 
end = 10
num = 100
numbers = [start+i*(end-start)/num for i in range(num)]
freqs = [norm.pdf(value, 5, 1) for value in numbers] #los argumentos de norm.pdf son numero de cusesos, media y desviacion estandar
plt.plot(numbers, freqs)
plt.grid(True)
plt.axvline(5,color="red")
plt.axvline(4,color="green")
plt.axvline(6,color="green")
plt.axvline(3,color="purple")
plt.axvline(7,color="purple")
plt.show()
```

```python
accumfreqs = [norm.cdf(value, 5, 1) for value in numbers] #los argumentos de norm.pdf  (Funcion Densidad Probabilidad) norm.cdf (Cumulative Density Function) son numero de cusesos, media y desviacion estandar
plt.plot(numbers, accumfreqs)
plt.grid(True)
plt.show()
```

```python
# una maquina produce sacos de centeno que pesan 3 kilos con una desviacin estndar de 0.5 kilos 
# un operador opina que la maquina no esta bien calibrada, toma unas muestras de sacos al azar [2.3, 2.9, 3.1, 2.5, 2.8, 3.0, 2.7] 
# la muestra indica que la mquina est dando valores diferentes a 3 kilos
# tiene razon el operador?
#H0: mu=3
#Ha= mu!=3
#nivel de significancia alfa = 0.05
sacks = np.array([2.3, 2.9, 3.1, 2.5, 2.8, 3.0, 2.7])
sample_mean = sacks.mean()
sample_std = sacks.std()
population_mean=3
population_std= 0.5
```

```python
zscore = (sample_mean-population_mean)/(population_std/np.sqrt(len(sacks)))# numerador es la diferencia entre la media muestral y la media poblacional, el denominador es la division entre la desviacion estandar poblacional y el tamao de la muestra
#zscore cuanto a la izquiera cae en el error estandar
pvalue = 2*(1-norm.cdf(abs(zscore)))
pvalue 
#Pvalue > alfa: no se rechaza la Hipotesis nula Ho
#Pvalue menor alfa: se acepta la hipotesis alterna. 
```
